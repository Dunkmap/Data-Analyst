#2nd Practice of handling missing values

```python
import pandas as pd
from io import StringIO

data = """EmployeeID,Name,Age,Department,Salary,JoinDate
101,John Smith,29,Sales,50000,2018-06-01
102,Emily Davis,?,Marketing,54000,2019-08-12
103,Michael Lee,,Finance,,2020-01-15
104,Ashley Brown,33,HR,48000,
105,Chris Green,27,,52000,2021-03-10
106,,45,IT,NaN,2017-11-23
107,Rachel Adams,31,Marketing,57000,?"""
ddf= pd.read_csv(StringIO(data))
ddf.info()
ddf.isnull().sum()
```

```python
ddf.replace('?', pd.NA, inplace=True)
ddf['Age'] = pd.to_numeric(ddf['Age'], errors='coerce')  # Convert 'Age' to numeric, coercing errors to NaN
ddf['JoinDate'] = pd.to_datetime(ddf['JoinDate'], errors='coerce')
ddf
```

```python
#convert the value of age by using mean method
ddf['Age'].fillna(ddf['Age'].mean(),inplace=True)
ddf
```

```python
import seaborn as sns
sns.histplot(ddf['Age'],kde=True)
```
