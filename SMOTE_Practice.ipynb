#SMOTE


```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples=1000, n_features=2, n_redundant=0,n_clusters_per_class=1, weights=[0.90],random_state=12)
```

```python
import pandas as pd
df1=pd.DataFrame(x, columns=['info1','info2'])
df2=pd.DataFrame(y,columns=['target'])
final_df=pd.concat([df1,df2], axis=1)
```

```python
from matplotlib import pyplot as plt
plt.scatter(final_df['info1'],final_df['info2'],c=final_df['target'])
```

```python
from imblearn.over_sampling import SMOTE
```

```python
oversample=SMOTE()
x,y= oversample.fit_resample(x,y)
```

```python
df1 = pd.DataFrame(x, columns=['info1','info2'])
df2 = pd.DataFrame(y, columns=['target'])
oversampled_df = pd.concat([df1, df2], axis=1)
```

```python

# Ensure oversampled_df is defined
if 'oversampled_df' not in globals():
    df1 = pd.DataFrame(x, columns=['info1','info2'])
    df2 = pd.DataFrame(y, columns=['target'])
    oversampled_df = pd.concat([df1, df2], axis=1)

# Scatter plot
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    oversampled_df['info1'],
    oversampled_df['info2'],
    c=oversampled_df['target'],
    cmap='coolwarm',
    alpha=0.7
)

# Class counts
count_0 = sum(oversampled_df['target'] == 0)
count_1 = sum(oversampled_df['target'] == 1)

# Add text annotation inside plot
plt.text(
    oversampled_df['info1'].min(),
    oversampled_df['info2'].max(),
    f'Class 0: {count_0}\nClass 1: {count_1}',
    fontsize=12,
    bbox=dict(facecolor='white', edgecolor='black')
)

# Labels & Title
plt.xlabel('info1')
plt.ylabel('info2')
plt.title('Scatter Plot with Class Counts')
plt.colorbar(scatter, label='Class Label')
plt.grid(True)
plt.show()

```
